### Task 1:
Suppose a TCS delivery boy has to deliver parcels from Head Office (MM Alam Road) to 5 different locations in Lahore (Johar Town, Shahdara, DHA Phase 6, Wapda Town, Allama Iqbal Town) and then return back to the Head Office. He wants to find the route with least travelling distance. Can you help him with that using Hill Climbing Algorithm? 

[You can construct distance matrix using google maps OR you can take random values (between 0-50) for distances between any two spots.]

```python
#Implement Simple HCS
dmap = {'A':[['B',8],['C',20]], 
            'B':[['D',6],['C',20],['A',25]], 
            'C':[['B',8],['E',12],['A',25]], 
            'D':[['F',0],['B',8],['E',12]],
            'E':[['F',0],['D',6],['C',20]],
            'F': [['D',6],['E',12]]
           }

Closed = list()



def MOVEGEN(N):
    New_list=list()
    if N in dmap.keys():
        New_list = dmap[N]
 
    return New_list

def SORT(L):
    L.sort(key = lambda x: x[1]) 
    return L 
 
def heu(Node):
    return Node[1]

def APPEND(L1,L2):
    New_list = list(L1) + list(L2)
    return New_list

def Hill_Climbing(Start):
    global Closed
    N = Start
    CHILD = MOVEGEN(N)
    SORT(CHILD)
    N = [Start,25]
    print("\nStart =",N)
    print("Sorted Child List =",CHILD)
    newNode = CHILD[0]
    CLOSED = [N]
 
    while heu(newNode) <= heu(N):
        print("\n---------------------")
        N = newNode
        print("N =",N)
        CLOSED = APPEND(CLOSED,[N])
        CHILD = MOVEGEN(N[0])
        SORT(CHILD)
        print("Sorted Child List =",CHILD)
        print("CLOSED =",CLOSED)
        newNode=CHILD[0]
 
    Closed = CLOSED
    
Hill_Climbing('A')
```
