<b>Apply Depth First Search (BFS) to traverse the following graph. Start your traversal from vertex S, and print down the order in which vertices will be visited during the traversal. And also print the total cost of traversal from S to G.</b>

```python
# Implement BFS Here
MyGraph = {
    'S' : ['A','C','I','K','M'],
    'A' : ['S'],
    'C' : ['H','S'],
    'E' : ['K','G'],
    'G' : ['E','I','P'],
    'H' : ['C'],
    'I' : ['G','K','S'],
    'K' : ['E','I','S'],
    'M' : ['P','S'],
    'P' : ['G','M']
}

def bfs(visited, MyGraph, node):
    queue = []
    visited.append(node)
    queue.append(node)

    while queue:
        s = queue.pop(0) 
        print (s, end = " ") 

        for n in MyGraph[s]:
            if n not in visited:
                visited.append(n)
                queue.append(n)

# Driver Code
bfs([], MyGraph, 'S')
```
